# Inferential Statistics

Calculate the 95% Confidence Interval (CI) for the age of people with diabetes

```python
import numpy as np
from scipy import stats

# ages is a list or numpy array of ages of people with diabetes
ages = df[df['Outcome']==1]['Age'].values

# Calculate the mean and standard deviation of the ages
mean = np.mean(ages)
std_dev = np.std(ages)

# Calculate the 95% Confidence Interval
ci = stats.norm.interval(0.95, loc=mean, scale=std_dev)

print(f"The 95% Confidence Interval is {ci}")
```

From report National Institutes of Health, mean BMI in patients with diabets was higher than (27.06 +- 3.6) than mean BMI in healthy patient(22.47 +- 0.99). So I want double with my data.

Link: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC10485520/#:~:text=Mean%20BMI%20in%20patients%20with,and%20BMI%20%5BTable%201%5D.

Null Hypothesis: The mean BMI in patients with diabetes is equal 27.06
Alternative Hypothesis: The mean BMI in patients with diabetes is higher than 27.06

95% interval for the age of people with diabetes is between 15 and 56 years old.

```python
from scipy import stats

# Replace this with your actual data
bmi_diabetes = np.array(df[df['Outcome'] == 1]['BMI'])

# Perform the one-sample t-test
t_stat, p_value = stats.ttest_1samp(bmi_diabetes, 27.06)

print(f"The t-statistic is {t_stat} and the p-value is {p_value}")
```

In conclusion, because the p-value is less than 0.05, we reject the null hypothesis and conclude that the mean BMI in patients with diabetes is higher than 27.06.

Null Hypothesis (H0): There is no significant difference in mean glucose levels between diabetic and non-diabetic individuals.

Alternative Hypothesis (H1): There is a significant difference in mean glucose levels between diabetic and non-diabetic individuals.

```python
from scipy import stats

diabetic_glucose = df[df['Outcome'] == 1]['Glucose']
non_diabetic_glucose = df[df['Outcome'] == 0]['Glucose']

t_stat, p_value = stats.ttest_ind(diabetic_glucose, non_diabetic_glucose)

print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

With P-value less then 0.05, we reject the null hypothesis and conclude that there is a significant difference in mean glucose levels between diabetic and non-diabetic individuals.
From that, we can infer that glucose levels are a significant predictor of diabetes.
